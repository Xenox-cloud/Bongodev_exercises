```python
 # Solution 1: Array creation and modification
import numpy as np

#Original code
array = np.random.randint(10,50, size=(4,4))
print("Original Array:\n", array)

# Modify the array
array[array > 40] = 0
print("Modified Array:\n", array)
```

```python
# Solution 2: Statistical Operations
# Task: Modify the code to compute the median and standard deviation instead of the mean, and 
# apply it to a 1D array of 20 random floats between 0 and 10
array = np.random.uniform(0, 10, size=20)
median_value = np.median(array)
std_deviation = np.std(array)
print("Array:", array)
print("Median: ", median_value)
print("Standard Deviation: ", std_deviation)
```

```python
# Solution 3: DataFrame Filtering
import pandas as pd
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'age': [25, 30, 35, 20],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago']
    }

df = pd.DataFrame(data)
filtered_df = df[(df['age'] > 25) & (df['city'] == 'New York')]
print("Filtered DataFrame:\n", filtered_df)
```

```python
# Solution 4: Grouping and Aggregation
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 20, 28],
    'city': ['New York', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles']
}

df = pd.DataFrame(data)
grouped = df.groupby('city')['age'].max()
print("Grouped max ages:\n", grouped)
```

```python
# Solution 5 Boxplot Customization
from pyparsing import col
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(
    style="darkgrid",
    palette='flare',
    font='sans-serif',
    font_scale=1
    )
# print(sns.get_dataset_names())
#original code
tips = sns.load_dataset('tips')
print(tips.head())
plot= sns.boxplot(
    data=tips, 
    x='day', 
    y='total_bill',
    hue='sex',
    )
plot.set(
    title='Total Bill Distribution'
)

plt.show()
```

```python
# Solution 6: Correlation Heatmap
tips = sns.load_dataset('tips')
sns.heatmap(tips.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
#solution 7: Interactive scatter plot
import plotly.express as px

tips = px.data.tips()
fig = px.scatter(
    tips, 
    x='total_bill', 
    y='tip', 
    color='day',
    title='Total Bill vs Tip',
    size='size',
    hover_data=['time', 'smoker']
    
    )
fig.show()
```

```python
# Solution 8: Barplot with customization
tips = px.data.tips().groupby(['day','sex'])['total_bill'].mean().reset_index()
fig = px.bar(
    tips,
    x='day',
    y='total_bill',
    color='sex',
)
fig.update_layout(
    title='Average Total Bill by Day and Sex')
fig.show()
```
